```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset

df = pd.read_csv('fintech_users.csv')

# Preview data
print(df.head())
```

```python
# Count retained users by signup month
retention = df.groupby('signup_month')['user_id'].nunique()

print(retention)
```

```python
retention.plot(kind='bar')

plt.title('User Retention by Signup Cohort')
plt.xlabel('Signup Month')
plt.ylabel('Retained Users')

plt.show()
```

```python
revenue = df.groupby('active_month')['revenue'].sum()

print(revenue)
```

```python
revenue.plot(kind='line', marker='o')

plt.title('Monthly Revenue Trends')
plt.xlabel('Active Month')
plt.ylabel('Revenue')

plt.show()
```

# Key Insights

- January and February cohorts demonstrated the strongest user retention
- Revenue activity declined as customer engagement decreased over time
- Higher transaction frequency was associated with stronger revenue performance
- Cohort analysis identified early customer drop-off trends across later signup groups
